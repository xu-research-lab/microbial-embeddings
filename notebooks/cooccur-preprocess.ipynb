### the times of co-occur

```python
times_value = []
for i in range(50):
    times_value.append(np.sum((cooccur_table['russell_rao'] > (i + 0.1) /n) & (cooccur_table['russell_rao'] < (i + 1.1)/n)) / len(cooccur_table['russell_rao']) * 100)
```

```python
plt.figure(figsize=(20, 15))
fig, ax1 = plt.subplots()
ax1.bar(np.arange(1, 51), times_value, label = "Percentage")
ax1.set_xlabel("The times of co-occur")
ax1.set_ylabel("Percentage")

ax2 = ax1.twinx()
ax2.plot(np.arange(1, 51), np.cumsum(times_value, dtype=float), '-*r', label = "Cumulative percentage")
ax2.set_ylabel("Cumulative percentage")
 
fig.legend(loc="upper right", bbox_to_anchor=(1, 0.2), bbox_transform=ax1.transAxes)
plt.show()
```

```python
times_index = {}
for i in range(21943):
    times_index[i] = np.argwhere((russell_rao.data > (i + 0.1) /n) & (russell_rao.data < (i + 1.1)/n)).flatten()
```

```python
times_value = []
for i in range(21943):
    times_value.append(len(times_index[i]))
```

```python
plt.figure(figsize=(60, 30))
plt.scatter(np.arange(1, 21944), times_value, s=2)
plt.yscale("log")
plt.tick_params(labelsize=30)
plt.xticks(np.arange(1, 21944, 1000))
plt.xlabel("The times of co-occur", size = 30)
plt.ylabel("The number of co-occur features", size = 30)
plt.show()
```

```python
for j in formula_list:   
    data = np.array(cooccur_table[j])
    mean_value = []
    sem_value = []
    for i in range(21943):
        tmp = data[times_index[i]]
        mean_value.append(np.mean(tmp))
        sem_value.append(np.std(tmp) / np.sqrt(len(tmp)))
    err_value = [float(i) * 1.96 for i in sem_value]
    plt.figure(figsize=(60, 30))
    plt.errorbar(np.arange(1, 21944), mean_value, yerr=err_value, fmt='o',ecolor='r',color='b',elinewidth=1,errorevery=10)
    plt.tick_params(labelsize=30)
    plt.xlabel("The times of co-occur", size = 30)
    plt.ylabel("Mean of co-occur", size = 30)
    plt.title(j, size = 30)
    plt.show()
```
